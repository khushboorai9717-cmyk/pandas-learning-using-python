```python
import pandas as pd
data = {
    "Name":["Sita", "Gita"," Radha"],
    "Age":[20,23,24,],
    "City":["Noida", "Delta", "Parichowk"],
    "salary":[2300,2400,2500],
    "performance score":[34,35,36]
}
df = pd.DataFrame(data)
print(df)
high_salary = df[df['salary']>5000]
print('Employee with salary>5000')
print(high_salary)
filtered = df[(df['Age']>30)&(df['salary']>500000)]
print(f"Employee list age > 30 + Salary>500000")
print(filtered)

filtered_or = df[(df['Age'] >35 )| (df['performance score']> 90)]
print(filtered_or)
```
